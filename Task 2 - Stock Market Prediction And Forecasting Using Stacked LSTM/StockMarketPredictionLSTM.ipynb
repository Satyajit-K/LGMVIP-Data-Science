```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler(feature_range = (0, 1))
df_close = scaler.fit_transform(np.array(df_close).reshape(-1, 1))
```

## **Train** **test** **split**

```python
training_size = int(len(df_close)*0.75)
test_size = len(df_close) - training_size
train_data, test_data = df_close[0:training_size,:], df_close[training_size: len(df_close), :1]
```

```python
training_size, test_size
train_data, test_data
```

### DATA PREPROCESSING

```python
def create_dataset(dataset, time_step = 1):
  dataX, dataY = [], []
  for i in range(len(dataset) - time_step - 1):
    a = dataset[i : (i+time_step), 0]
    dataX.append(a)
    dataY.append(dataset[i+time_step, 0])
  return np.array(dataX), np.array(dataY)
```

```python
time_step = 100
trainX, trainY = create_dataset(train_data, time_step)
testX, testY = create_dataset(test_data, time_step)
```

## **LSTM**

```python
trainX = trainX.reshape(trainX.shape[0], trainX.shape[1], 1)
testX = testX.reshape(testX.shape[0], testX.shape[1], 1)
```

```python
print(trainX.shape)
print(trainY.shape)
```

```python
print(testX.shape)
print(testY.shape)
```

# **MODULES** **FOR** **THE** **STACKED** **LSTM**

```python
import math
from sklearn.metrics import mean_squared_error
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.layers import LSTM
```

```python
tf.__version__
```

```python
model = Sequential()
model.add(LSTM(50, return_sequences = True, input_shape = (100, 1)))
model.add(LSTM(50, return_sequences = True))
model.add(LSTM(50))
model.add(Dense(1))
model.compile(loss = 'mean_squared_error', optimizer = 'adam')
```

```python
model.summary
```

```python
model.fit(trainX, trainY, validation_data = (testX, testY), epochs = 100, batch_size = 64, verbose = 1)
```

```python
train_predictions = model.predict(trainX)
test_predictions = model.predict(testX)
```

```python
train_predictions = scaler.inverse_transform(train_predictions)
test_predictions = scaler.inverse_transform(test_predictions)
```

```python
math.sqrt(mean_squared_error(trainY, train_predictions))
```
